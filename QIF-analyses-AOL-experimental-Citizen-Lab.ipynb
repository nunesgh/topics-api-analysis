### Experiments 4 and 5: Topics with Generalization and Bounded Noise on AOL experimental dataset with Citizen Lab Classification; and Topics with Generalization, Bounded Noise, and Differential Privacy on AOL experimental dataset with Citizen Lab Classification

```python
display(aol_experimental_citizen_lab_generalization)
```

```python
%%time
aol_experimental_citizen_lab_seen_topics = []
for tup in aol_experimental_citizen_lab_generalization.itertuples():
    aol_experimental_citizen_lab_seen_topics.extend(tup.AllTopics.replace('[','').replace(']','').replace(' ','').replace('\'','').replace('"','').split(','))
aol_experimental_citizen_lab_seen_topics = list(set(aol_experimental_citizen_lab_seen_topics))
display(len(aol_experimental_citizen_lab_seen_topics))
```

#### Privacy

```python
temp_top_S = {tup.Index : {t : 1 for t in tup.sTopics.replace('[','').replace(']','').replace(' ','').replace('\'','').replace('"','').split(',')} for tup in aol_experimental_citizen_lab_generalization.itertuples()}
```

```python
%%time
S = 5
M = len(aol_experimental_citizen_lab_seen_topics)
R = 0.05

aol_experimental_citizen_lab_channel_bn = pandas.DataFrame(data=temp_top_S).T.applymap(lambda x: 1/S, na_action='ignore').fillna(0)
aol_experimental_citizen_lab_channel_bn = aol_experimental_citizen_lab_channel_bn[sorted(aol_experimental_citizen_lab_channel_bn.columns.to_list())]
aol_experimental_citizen_lab_channel_bn = aol_experimental_citizen_lab_channel_bn.sort_index()

aol_experimental_citizen_lab_channel_bn_dp = pandas.DataFrame(data=temp_top_S).T.applymap(lambda x: ((1-R)/S) + (R/M), na_action='ignore').fillna(R/M)
aol_experimental_citizen_lab_channel_bn_dp = aol_experimental_citizen_lab_channel_bn_dp[sorted(aol_experimental_citizen_lab_channel_bn_dp.columns.to_list())]
aol_experimental_citizen_lab_channel_bn_dp = aol_experimental_citizen_lab_channel_bn_dp.sort_index()
```

```python
display(aol_experimental_citizen_lab_channel_bn.cumsum(axis=1))
display(aol_experimental_citizen_lab_channel_bn_dp.cumsum(axis=1))
```

```python
%%time
N = len(aol_experimental_citizen_lab_channel_bn.index)
aol_experimental_citizen_lab_bn_dp_prior = numpy.array([1/N for row in range(N)])
```

```python
%%time
display("Prior: " + str(qif.measure.bayes_vuln.prior(aol_experimental_citizen_lab_bn_dp_prior)))
display("BN Posterior: " + str(qif.measure.bayes_vuln.posterior(aol_experimental_citizen_lab_bn_dp_prior, aol_experimental_citizen_lab_channel_bn.to_numpy())))
display("DP Posterior: " + str(qif.measure.bayes_vuln.posterior(aol_experimental_citizen_lab_bn_dp_prior, aol_experimental_citizen_lab_channel_bn_dp.to_numpy())))
```

```python
%%time
display("BN Leakage: " + str(qif.measure.bayes_vuln.posterior(aol_experimental_citizen_lab_bn_dp_prior, aol_experimental_citizen_lab_channel_bn.to_numpy())/qif.measure.bayes_vuln.prior(aol_experimental_citizen_lab_bn_dp_prior)))
display("DP Leakage: " + str(qif.measure.bayes_vuln.posterior(aol_experimental_citizen_lab_bn_dp_prior, aol_experimental_citizen_lab_channel_bn_dp.to_numpy())/qif.measure.bayes_vuln.prior(aol_experimental_citizen_lab_bn_dp_prior)))
```

#### Utility

```python
%%time
aol_experimental_citizen_lab_top_S = []
for k,v in temp_top_S.items():
    temp = str(list(v.keys()))
    temp = temp.replace("\'","")
    aol_experimental_citizen_lab_top_S.append(temp)

aol_experimental_citizen_lab_top_S_count = 0
aol_experimental_citizen_lab_top_S_counts = dict()
for case in aol_experimental_citizen_lab_top_S:
    if case in aol_experimental_citizen_lab_top_S_counts:
        aol_experimental_citizen_lab_top_S_counts[case] = aol_experimental_citizen_lab_top_S_counts[case] + 1
    else:
        aol_experimental_citizen_lab_top_S_counts[case] = 1
    aol_experimental_citizen_lab_top_S_count += 1

aol_experimental_citizen_lab_top_S_prior = aol_experimental_citizen_lab_top_S_counts
for case in aol_experimental_citizen_lab_top_S_prior:
    aol_experimental_citizen_lab_top_S_prior[case] = aol_experimental_citizen_lab_top_S_prior[case] / aol_experimental_citizen_lab_top_S_count

d = {'topics' : [], 'prior' : []}
for k,v in aol_experimental_citizen_lab_top_S_prior.items():
    d['topics'].append(k)
    d['prior'].append(v)
aol_experimental_citizen_lab_top_S_prior = pandas.DataFrame(data=d)
aol_experimental_citizen_lab_top_S_prior = aol_experimental_citizen_lab_top_S_prior.set_index('topics').sort_index()
```

```python
display(aol_experimental_citizen_lab_top_S_prior)
```

```python
%%time
S = 5
M = len(aol_experimental_citizen_lab_seen_topics)
R = 0.05

aol_experimental_citizen_lab_channel_bn = pandas.DataFrame(columns=[c for c in aol_experimental_citizen_lab_seen_topics], index=d['topics'])
aol_experimental_citizen_lab_channel_bn_dp = pandas.DataFrame(columns=[c for c in aol_experimental_citizen_lab_seen_topics], index=d['topics'])
for tup in aol_experimental_citizen_lab_channel_bn.itertuples():
    temp = tup.Index.replace('[','').replace(']','').split(', ')
    for col in aol_experimental_citizen_lab_channel_bn.columns:
        aol_experimental_citizen_lab_channel_bn.at[tup.Index, col] = 1/S if col in temp else 0
        aol_experimental_citizen_lab_channel_bn_dp.at[tup.Index, col] = ((1-R)/S) + (R/M) if col in temp else (R/M)

aol_experimental_citizen_lab_channel_bn = aol_experimental_citizen_lab_channel_bn[sorted(aol_experimental_citizen_lab_channel_bn.columns.to_list())]
aol_experimental_citizen_lab_channel_bn = aol_experimental_citizen_lab_channel_bn.sort_index()

aol_experimental_citizen_lab_channel_bn_dp = aol_experimental_citizen_lab_channel_bn_dp[sorted(aol_experimental_citizen_lab_channel_bn_dp.columns.to_list())]
aol_experimental_citizen_lab_channel_bn_dp = aol_experimental_citizen_lab_channel_bn_dp.sort_index()
```

```python
display(aol_experimental_citizen_lab_channel_bn.cumsum(axis=1))
display(aol_experimental_citizen_lab_channel_bn_dp.cumsum(axis=1))
```

```python
%%time
display("Prior: " + str(qif.measure.bayes_vuln.prior(aol_experimental_citizen_lab_top_S_prior['prior'].to_numpy())))
display("BN Posterior: " + str(qif.measure.bayes_vuln.posterior(aol_experimental_citizen_lab_top_S_prior['prior'].to_numpy(), aol_experimental_citizen_lab_channel_bn.to_numpy())))
display("DP Posterior: " + str(qif.measure.bayes_vuln.posterior(aol_experimental_citizen_lab_top_S_prior['prior'].to_numpy(), aol_experimental_citizen_lab_channel_bn_dp.to_numpy())))
```

```python
%%time
display("BN Leakage: " + str(qif.measure.bayes_vuln.posterior(aol_experimental_citizen_lab_top_S_prior['prior'].to_numpy(), aol_experimental_citizen_lab_channel_bn.to_numpy())/qif.measure.bayes_vuln.prior(aol_experimental_citizen_lab_top_S_prior['prior'].to_numpy())))
display("DP Leakage: " + str(qif.measure.bayes_vuln.posterior(aol_experimental_citizen_lab_top_S_prior['prior'].to_numpy(), aol_experimental_citizen_lab_channel_bn_dp.to_numpy())/qif.measure.bayes_vuln.prior(aol_experimental_citizen_lab_top_S_prior['prior'].to_numpy())))
```

**IBA gain**

```python
%%time
G_IBA = pandas.DataFrame(columns=[c for c in aol_experimental_citizen_lab_seen_topics], index=d['topics'])
for tup in G_IBA.itertuples():
    temp = tup.Index.replace('[','').replace(']','').split(', ')
    for col in G_IBA.columns:
        G_IBA.at[tup.Index, col] = 1 if col in temp else 0
G_IBA = G_IBA[sorted(G_IBA.columns.to_list())]
G_IBA = G_IBA.sort_index()
display(G_IBA.T)
```
